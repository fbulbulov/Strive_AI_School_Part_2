## Annotating images

### Rectangles

```python
# Draw green rectangle on top of a region of interest in your image
#dup=img.copy()
green_ret=cv2.rectangle(dup,(175,90),(320,330),(0,255,0),2)
```

```python
green_ret=cv2.rectangle(dup,(315,52),(480,330),(0,0,255),2)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(cv2.cvtColor(green_ret,cv2.COLOR_BGR2RGB))
plt.show
```

[Hint](# "You will need to use the cv2 function rectangle(img, pt1, pt2, color, thickness) -> img") 

### Adding text to images

```python
# Make a copy of your image and then write some text on top of the image
nature = green_ret.copy()
color = (0, 204, 255)

cv2.putText(nature, 'Nature', (10,80),cv2.FONT_HERSHEY_TRIPLEX , 1, color, 1);
```

```python
plt.figure(figsize=(15,15))
plt.imshow(cv2.cvtColor(nature,cv2.COLOR_BGR2RGB))
plt.show
```

[Hint](# "You will need to use cv2 putText function -> putText(img, text, coordinates, fontFace, fontScale, color, thickness, lineType). For fonts check the enum cv2.FONT_HERSHEY_")

## Level 1 and Level 2 QR Detection

```python
img = cv2.imread('info.png')

detector=cv2.QRCodeDetector()

data, bbox, straight_qrcode=detector.detectAndDecode(img)

if bbox is not  None:
    print(data)
    n_lines=len(bbox)
    for i in range(n_lines):
        point1=(25, 25)
        point2=(225, 225)
        #point1=tuple(bbox[i][0])
        #point2 = tuple(bbox[(i+1) %n_lines][0])
        cv2.rectangle(img, point1, point2, color=(0,255,0), thickness=2)

#displaying the results
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
