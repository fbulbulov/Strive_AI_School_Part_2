**Part 1: conversion of the handwritings**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def imshow(img):
    plt.figure(figsize=(15,15))
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB), cmap='gray')
```

```python
mnist_styled=cv2.imread('img/4.png')
imshow(mnist_styled)
```

```python
img=cv2.imread('img/digits1.jpg')
imshow(img)
```

```python
#Bluring
img_copy=img.copy()
denoiced = cv2.fastNlMeansDenoisingColored(img_copy,None,10,10,7,21) 
imshow(denoiced)
```

```python
#additional filters
grayscale_img=cv2.cvtColor(denoiced, cv2.COLOR_BGR2GRAY)
thresh =cv2.adaptiveThreshold(grayscale_img,
                             255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                             cv2.THRESH_BINARY_INV,23,8)
plt.figure(figsize=(20,15))
plt.imshow(thresh,cmap='gray')
```

Erosion/ dilate

kernel=cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3))
kernel

```python
erosion=cv2.dilate(thresh,kernel, iterations=4)
plt.figure(figsize=(20,15))
plt.figure(figsize=(20,15))
plt.imshow(erosion,cmap='gray')
```

We wiill thresh version, as it look closer to MNIST Dataset

```python
plt.imshow(thresh,cmap='gray')
```

Cropping, resizing and saving the image into the separable digits

```python
#resizer
def resize(image):
    w,h = 28,28
    return cv2.resize(image,(w,h))
```

```python
# 1
one = thresh[190:310, 120:240]
one = resize(one)
plt.figure(figsize=(5,5))
plt.imshow(one,cmap='gray')
print(one.shape)
cv2.imwrite('f_mnist//1.png',one)
```

```python
#3
three= thresh[190:310, 225:345]
three = resize(three)
plt.figure(figsize=(5,5))
plt.imshow(three,cmap='gray')
print(three.shape)
cv2.imwrite('f_mnist//3.png',three)
```

```python
#5
five= thresh[200:320, 350:470]
five = resize(five)
plt.figure(figsize=(5,5))
plt.imshow(five,cmap='gray')
print(five.shape)
cv2.imwrite('f_mnist//5.png',five)
```

```python
#four
four= thresh[210:330, 440:560]
four = resize(four)
plt.figure(figsize=(5,5))
plt.imshow(four,cmap='gray')
print(four.shape)
cv2.imwrite('f_mnist//4.png',four)
```

**Part 2: Image classificatons / predictions**

```python
import torch
import torchvision
```

```python
from torchvision import datasets, transforms
root_dir = 'f_mnist/'
transformer = transforms.Compose([                                      
                                       transforms.ToTensor(),
                                       transforms.Normalize([0.5, 0.5, 0.5],
                                                            [0.5, 0.5, 0.5])])
# Pass transforms in here, then run the next cell to see how the transforms look
test_dataset = datasets.ImageFolder(root_dir, transform=transformer)
```

```python
len(test_dataset)
```

```python
img, label = test_dataset[2]
plt.imshow(img[2], cmap='gray')
print('Shape:', img.shape)
#print('Label:', label)
```

```python
model = torch.load('./mnist-logistic.pth')
model
```

```python

model.load_state_dict(torch.load('mnist-logistic.pth'))
```

```python
def predict_image(img, model):
    xb = img.unsqueeze(0)
    yb = model(xb)
    _, preds = torch.max(yb, dim=1)
    return preds[0].item()
```

```python
img, label = test_dataset[0]
plt.imshow(img[0], cmap='gray')
print('Label:', label, ', Predicted:', predict_image(img, model))
```
