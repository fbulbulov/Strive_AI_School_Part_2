```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
print(f'OpenCV: {cv2.__version__}')
```

```python
def imshow(img):
    #if enlarge:
       # plt.figure(figsize=(20,15));
   # plt.imshow(img[:,:,::-1])
    plt.figure(figsize=(15,15))
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
gray_scene= cv2.imread('img/WandoScene.jpg',cv2.IMREAD_GRAYSCALE)
gray_template = cv2.imread('img/waldo.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
match=cv2.matchTemplate(gray_scene, gray_template, cv2.TM_SQDIFF)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(match)
top_left=max_loc
```

```python
bottom_right=(top_left[0])
```

```python
#Reading anc conversion of the pictures to grayscale
img= cv2.imread('img/WaldoScene1.jpg')
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template = cv2.imread('img/waldo.jpg', cv2.IMREAD_GRAYSCALE)
#template_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('template',template)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
result=cv2.matchTemplate(gray_img, template, cv2.TM_CCOEFF)
```

```python
cv2.imshow('img',img)
cv2.imshow('result',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#printing result
print (result)
```
