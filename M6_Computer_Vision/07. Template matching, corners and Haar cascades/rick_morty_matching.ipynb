```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
print(f'OpenCV: {cv2.__version__}')
```

```python
def imshow(img):
    #if enlarge:
       # plt.figure(figsize=(20,15));
   # plt.imshow(img[:,:,::-1])
    plt.figure(figsize=(15,15))
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
#Reading anc conversion of the pictures to grayscale
img= cv2.imread('img/main.jpg')
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template = cv2.imread('img/rick.jpg', cv2.IMREAD_GRAYSCALE)
#template_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('template',template)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#matching the main picture with the template
result=cv2.matchTemplate(gray_img, template, cv2.TM_CCOEFF_NORMED)
```

```python
cv2.imshow('img',img)
cv2.imshow('result',result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#printing result
print (result)
```

```python
# finding where it is close to 0.8
loc=np.where(result >= 0.99)
```

```python
print (loc)
```

```python
for pt in zip(*loc):
    print(pt)
    
```

```python
#getting second point of the template size
w, h= template.shape[::-1]
```

```python
for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0]+w,pt[1]+h), (0,255,0), 3)

cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
