```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
print(f'OpenCV: {cv2.__version__}')
```

```python
def imshow(img):
    #if enlarge:
       # plt.figure(figsize=(20,15));
   # plt.imshow(img[:,:,::-1])
    plt.figure(figsize=(15,15))
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
gray_scene= cv2.imread('img/WandoScene.jpg',cv2.IMREAD_GRAYSCALE)
gray_template = cv2.imread('img/waldo.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
match=cv2.matchTemplate(gray_scene, gray_template, cv2.TM_CCOEFF_NORMED)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(match)
top_left=max_loc
bottom_right=min_loc
```

```python
bottom_right=(top_left[0]+50, top_left[1]+50)
cv2.rectangle(gray_scene,top_left,bottom_right,(0,255,0),2)
imshow(gray_scene)
```
