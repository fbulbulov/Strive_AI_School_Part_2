```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
print(f'OpenCV: {cv2.__version__}')
```

```python
def imshow(img):
    #if enlarge:
       # plt.figure(figsize=(20,15));
   # plt.imshow(img[:,:,::-1])
    plt.figure(figsize=(15,15))
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

### Template Matching

```python
# Load img/WandoScene.jpg  as the main image
# Load img/waldo.jpg as the template
# Corvert the images to grayscale
# Find the template in the image
# Draw a rectagle around the the detected match
```

```python
scene= cv2.imread('img/WandoScene.jpg')
template = cv2.imread('img/waldo.jpg')
```

```python
gray_scene=cv2.cvtColor(scene,cv2.COLOR_BGR2GRAY)
gray_template=cv2.cvtColor(template,cv2.COLOR_BGR2GRAY)
```

```python
match=cv2.matchTemplate(gray_scene, gray_template, cv2.TM_CCOEFF_NORMED)
```

```python
imshow(match)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(match)
top_left=max_loc
bottom_right=min_loc
```

```python
print(top_left,bottom_right)
```

```python
bottom_right=(top_left[0]+50, top_left[1]+50)
cv2.rectangle(scene,top_left,bottom_right,(255,0,0),2)
imshow(scene)
```
