```python
dataset=[]

for _ in range(10):

    for _,row in df.iterrows():
        temp={
            "age": row['age']+np.random.uniform(age_s),
            
            #"sex": round(row['sex']+np.random.uniform(sex_s),0),
            "sex": row['sex']+np.random.randint(0,1),
            
            "cp": row['cp']+np.random.uniform(cp_s),
            "trestbps": row['trestbps']+np.random.uniform(trestbps_s),
            "chol": row['chol']+np.random.uniform(chol_s),
            "fbs": row['fbs']+np.random.uniform(fbs_s),
            "restecg": row['restecg']+np.random.uniform(restecg_s),
            "thalach": row['thalach']+np.random.uniform(thalach_s),
            "exang": row['exang']+np.random.uniform(exang_s),
            "oldpeak": row['oldpeak']+np.random.uniform(oldpeak_s),
            "slope": row['slope']+np.random.uniform(oldpeak_s),
            "ca": row['ca']+np.random.uniform(ca_s),
            "thal": row['thal']+np.random.uniform(thal_s),
            "target": round(row['target']+np.random.uniform(target_s),0),
            "target": row['target']+np.random.randint(0,1),
            #"target": round(row['target']+np.random.uniform(target_s),0) #rounded
            

        }
        dataset.append(temp)
        

print(len(dataset),"augmented data created.")
df_au=pd.DataFrame(dataset)
df_au
```
