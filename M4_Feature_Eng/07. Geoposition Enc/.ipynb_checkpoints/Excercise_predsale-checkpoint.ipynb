## Exercise 2: Pysical vs Online shop

Create a binary column to determint if the shop is online or not.

```python
df_shop['shop_name']=df_shop['shop_name'].str.lower() #all lower case
df_shop['shop_name']=df_shop['shop_name'].str.replace(r'[^\w\d\s]', ' ') #replacing all symbols
df_shop.head()
```

Type of shops

```python
#defining the type of the shops
df_shop['shop_type']='regular' # all shops are defaulted as regular and creation of the new column

df_shop.loc[df_shop['shop_name'].str.contains(r'tc'),'shop_type'] = 'tc'
df_shop.loc[df_shop['shop_name'].str.contains(r'mall|center|mega'),'shop_type'] = 'mall'
df_shop.loc[df_shop.shop_id.isin([9,20]),'shop_type'] = 'special'
df_shop.loc[df_shop.shop_id.isin([12,55]), 'shop_type']='online'

#pie plotting the new column shop_type
df_shop['shop_type'].value_counts().plot.pie(autopct='%1.0f%%', label='')
plt.title('Shop Types')
plt.show()
```

```python
#Online and Physical shops
df_shop['shop_set']='physical'


df_shop.loc[df_shop['shop_type'].str.contains(r'online'),'shop_set']='online'

#Pie plotting

df_shop['shop_set'].value_counts().plot.pie(autopct='%1.0f%%', label='')
plt.title('Online/Physical shops')
plt.show()
```

## Exercise 3: Get the cities

Extract the city name of each shop. Cities apears on the first word of the shop name.

```python
#First word represents the city name and we adding it to the new column named shop_city
df_shop['shop_city']=df_shop['shop_name'].str.split().str[0]
df_shop.head(3)
```

```python
# 9, 12, 55 no-city detected
df_shop.loc[df_shop.shop_id.isin([9,12,55]), 'shop_city']='No-city'
```

```python
# Plotting the pie chart
df_shop['shop_city'].value_counts().plot.pie(label='', )
plt.show()
```
